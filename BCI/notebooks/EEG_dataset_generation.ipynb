```python
import pandas as pd
import numpy as np
import os
import sys
sys.path.append('../')
from utils.loader import load_data, unicorn_fs, unicorn_eeg_channels



path_recording = '../data/recordings/recording_11'
path_dataset = '../data/dataset/dataset_11'


#------------------------------------------------------------------------------------
labels = [ 'bpm_90', 'song_1', 'bpm_150', 'song_2']
recording_type = ['listening','playing']


'''
4 EEG recordings WITH OSCAR in all the possible combinations of:

- recording type: [playing]
- labels: [bpm_60, relax, bpm_120, excited]

A recording is done as the following:

1. 5 seconds to stabilize the signal 
2. 30 second baseline
3. 1 minute bpm 90
4. 30 second baseline
5. 1 minute song 1 (90 bpm)
6. 30 second baseline
7. 1 minute bpm 150
8. 30 second baseline
9. 1 minute song 2 (150 bpm)

total: 6 minutes and 5 seconds per recording

'''

# count number of directories in path_recording
n_takes = len([d for d in os.listdir(path_recording) if os.path.isdir(os.path.join(path_recording, d))])
print(f'Number of takes: {n_takes}')


for rec_type in recording_type:
    bpm_90_list = []
    song_1_list = []
    bpm_150_list = []
    song_2_list = []
    baseline_list = []
    for take in range(1, n_takes+1):
        # Path
        path_data = os.path.join(path_recording, 'take_'+str(take), rec_type)

        if not os.path.exists(path_data):
            print(f'Path does not exist:\t{path_data}')
            continue

        files = [f for f in os.listdir(path_data) if f.endswith('.csv')]
        path_file = os.path.join(path_data, files[0])

        # Load data
        skip_seconds = 5
        eeg, trigger, dataframe = load_data(path_file, header=False, fs=unicorn_fs, skiprows=skip_seconds, names = unicorn_eeg_channels)

        # Duration in minutes and seconds of the recording
        duration = eeg.shape[0] / unicorn_fs
        minutes = int(duration / 60)
        seconds = int(duration % 60)
        print(f'The recording number {take} is {minutes} minutes and {seconds} seconds long')

        print(f'Processing file: {path_file}')

        # 30 seconds silence
        baseline_list.append(eeg[:unicorn_fs*30, :])
        # 1 minute metronome at 60 bpm
        bpm_90_list.append(eeg[unicorn_fs*30:unicorn_fs*90, :])
        # 30 seconds silence
        baseline_list.append(eeg[unicorn_fs*90:unicorn_fs*120, :])
        # 1 minute metronome at 90 bpm
        song_1_list.append(eeg[unicorn_fs*120:unicorn_fs*180, :])
        # 30 seconds silence
        baseline_list.append(eeg[unicorn_fs*180:unicorn_fs*210, :])
        # 1 minute metronome at 120 bpm
        bpm_150_list.append(eeg[unicorn_fs*210:unicorn_fs*270, :])
        # 30 seconds silence
        baseline_list.append(eeg[unicorn_fs*270:unicorn_fs*300, :])
        # 1 minute metronome at 150 bpm
        song_2_list.append(eeg[unicorn_fs*300:unicorn_fs*360, :])

    classes_list = [bpm_90_list, song_1_list, bpm_150_list, song_2_list]

    for label, class_list in zip(labels, classes_list):
        # convert to dataframes
        class_flatten = np.concatenate(class_list)
        class_df = pd.DataFrame(class_flatten)
        # save dataframes
        class_file_name = label + '.csv'
        path_class_folder = os.path.join(path_dataset, rec_type, label)
        if not os.path.exists(path_class_folder):
            os.makedirs(path_class_folder)
        path_class = os.path.join(path_class_folder, class_file_name)
        class_df.to_csv(path_class, index=False, header=False)
        print(f'File saved in:\t{path_class}')

    # concatenate silences and save them as baseline
    baseline_flatten = np.concatenate(baseline_list)
    baseline_df = pd.DataFrame(baseline_flatten)
    baseline_file_name = 'baseline.csv'
    path_baseline_folder = os.path.join(path_dataset, rec_type, 'baseline')

    if not os.path.exists(path_baseline_folder):
        os.makedirs(path_baseline_folder)

    path_baseline = os.path.join(path_baseline_folder, baseline_file_name)
    baseline_df.to_csv(path_baseline, index=False, header=False)

    print(f'File saved in:\t{path_baseline}\n')
```
